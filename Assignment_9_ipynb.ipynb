Q21.What are the assumptions for applying the Central Limit Theorem?

Ans The Central Limit Theorem (CLT) is a foundational concept in statistics, but to apply it correctly, certain assumptions must be met. Here are the key assumptions:

1. Random Sampling

The data must be obtained through a random process, such as random sampling or random assignment.

Ensures the sample is representative of the population.

2. Independence

Observations in the sample must be independent of each other.

If sampling without replacement, the sample size should be no more than 10% of the population to approximate independence (the 10% condition).

3. Sample Size

The sample size should be sufficiently large:

n â‰¥ 30 is a common rule of thumb for most populations.

If the population distribution is strongly skewed or not normal, a larger n may be needed.

If the population is normally distributed, even a small sample size is acceptable.

 4. Finite Variance

The population must have a finite standard deviation (i.e., the variance cannot be infinite).

This ensures the sample mean will converge to a normal distribution.

```python
2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Parameters
low = 1      # Lower bound (inclusive)
high = 6     # Upper bound (inclusive)
size = 10000  # Number of samples

# Generate discrete uniform distribution samples
samples = np.random.randint(low, high + 1, size=size)

# Count frequencies
counts = Counter(samples)

# PMF: probability = count / total samples
x_vals = sorted(counts.keys())
pmf_vals = [counts[x] / size for x in x_vals]

# Plot PMF
plt.bar(x_vals, pmf_vals, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('PMF of Discrete Uniform Distribution')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
        x (int): The value (0 or 1)
        p (float): Probability of success (0 <= p <= 1)

    Returns:
        float: Probability P(X = x)
    """
    if x not in (0, 1):
        return 0
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")
    return p if x == 1 else 1 - p

# Example usage
print("P(X = 0):", bernoulli_pdf(0, 0.3))  # Should return 0.7
print("P(X = 1):", bernoulli_pdf(1, 0.3))  # Should return 0.3
```

```python
6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution.

import numpy as np
import matplotlib.pyplot as plt

# Define parameters of the discrete uniform distribution
a = 1   # lower bound (inclusive)
b = 6   # upper bound (inclusive)

# Create the range of possible values
x = np.arange(a, b + 1)

# Since it's uniform, all probabilities are equal
pmf = np.ones_like(x) / len(x)

# Calculate the CDF
cdf = np.cumsum(pmf)

# Plot the CDF
plt.step(x, cdf, where='post', label='CDF', color='blue', linewidth=2)
plt.scatter(x, cdf, color='red')  # Show discrete points

plt.title('CDF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.ylim(0, 1.1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
9.Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores as a scatter plot
    plt.figure(figsize=(10, 5))
    plt.scatter(range(len(z_scores)), z_scores, color='purple', alpha=0.6)
    plt.axhline(0, color='black', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='red', linestyle='--', label='Z=1')
    plt.axhline(-1, color='blue', linestyle='--', label='Z=-1')
    plt.title('Z-Scores of the Dataset')
    plt.xlabel('Data Point Index')
    plt.ylabel('Z-Score')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    return z_scores
```

```python
17.Generate random variables and calculate their corresponding probabilities using the binomial distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def binomial_distribution_demo(n=10, p=0.5, num_samples=10):
    # Step 1: Generate random variables from Binomial distribution
    random_vars = np.random.binomial(n=n, p=p, size=num_samples)
    print("Random variables:", random_vars)

    # Step 2: Calculate the probability of each outcome using PMF
    x_values = np.arange(0, n + 1)
    probabilities = binom.pmf(x_values, n=n, p=p)

    print("\nBinomial Probabilities (PMF):")
    for x, prob in zip(x_values, probabilities):
        print(f"P(X = {x}) = {prob:.4f}")

    # Step 3: Plot the PMF
    plt.figure(figsize=(10, 5))
    plt.bar(x_values, probabilities, color='skyblue', edgecolor='black')
    plt.title(f'Binomial Distribution PMF (n={n}, p={p})')
    plt.xlabel('Number of Successes (x)')
    plt.ylabel('Probability P(X = x)')
    plt.grid(True, axis='y')
    plt.show()

# Example usage
binomial_distribution_demo(n=10, p=0.5, num_samples=10)
```

```python
18.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """Calculate Z-score"""
    z = (x - mean) / std
    return z

def plot_z_score(z):
    """Plot the standard normal distribution and mark the Z-score"""
    x_values = np.linspace(-4, 4, 1000)
    y_values = norm.pdf(x_values)

    plt.figure(figsize=(10, 5))
    plt.plot(x_values, y_values, label='Standard Normal Distribution', color='blue')

    # Shade area up to the Z-score
    plt.fill_between(x_values, y_values, where=(x_values <= z), color='skyblue', alpha=0.5)

    # Draw vertical line at Z-score
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title('Z-Score on Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example: Given data point and population parameters
x = 75        # Data point
mean = 70     # Mean of the distribution
std = 5       # Standard deviation

# Calculate Z-score and compare
z = calculate_z_score(x, mean, std)
print(f"Z-score for x = {x}: {z:.2f}")

# Plot comparison with standard normal distribution
plot_z_score(z)
```

```python
19.Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Compute p-value (two-tailed)
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Decision
    if p_value < alpha:
        result = "Reject the null hypothesis (Hâ‚€)"
    else:
        result = "Fail to reject the null hypothesis (Hâ‚€)"

    # Output
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z:.2f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Decision: {result}")

# Example sample data
sample_data = [102, 98, 105, 101, 99, 100, 97, 103, 104, 100]

# Known population parameters
population_mean = 100
population_std = 3
alpha = 0.05

# Run Z-test
z_test(sample_data, population_mean, population_std, alpha)
```

```python
20.Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    # Calculate sample statistics
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation

    # Z-critical value for two-tailed test
    z_critical = norm.ppf(1 - (1 - confidence) / 2)

    # Margin of error
    margin_error = z_critical * (sample_std / np.sqrt(n))

    # Confidence interval
    lower_bound = sample_mean - margin_error
    upper_bound = sample_mean + margin_error

    # Output
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident that the true population mean lies within this interval.")

# Example dataset
data = [101, 98, 97, 100, 99, 102, 103, 100, 98, 99]

# Calculate and interpret 95% confidence interval
confidence_interval(data, confidence=0.95)
```

```python
21.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def generate_normal_data_and_ci(mean=50, std=10, size=100, confidence=0.95):
    # Step 1: Generate normally distributed data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Step 2: Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)

    # Step 3: Z-critical value
    z_critical = norm.ppf(1 - (1 - confidence) / 2)

    # Step 4: Margin of Error and Confidence Interval
    margin_error = z_critical * (sample_std / np.sqrt(n))
    ci_lower = sample_mean - margin_error
    ci_upper = sample_mean + margin_error

    # Step 5: Print and interpret result
    print(f"Generated sample size: {n}")
    print(f"Sample mean = {sample_mean:.2f}")
    print(f"Sample std deviation = {sample_std:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident that the true mean lies within this interval.")

    # Step 6: Plot data histogram with mean and CI
    plt.figure(figsize=(10, 5))
    plt.hist(data, bins=20, color='lightblue', edgecolor='black', alpha=0.7)
    plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
    plt.axvline(ci_lower, color='green', linestyle='--', label=f'CI Lower = {ci_lower:.2f}')
    plt.axvline(ci_upper, color='green', linestyle='--', label=f'CI Upper = {ci_upper:.2f}')
    plt.title('Histogram of Generated Normal Data with Confidence Interval')
    plt.xlabel('Data Values')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the function
generate_normal_data_and_ci(mean=50, std=10, size=100, confidence=0.95)
```

```python
23.Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lam=4, max_k=15):
    # Step 1: Generate x-values (discrete)
    x = np.arange(0, max_k + 1)

    # Step 2: Calculate CDF values for each x
    cdf_values = poisson.cdf(x, mu=lam)

    # Step 3: Print CDF values
    print(f"Cumulative Distribution Function (CDF) for Poisson(Î»={lam}):")
    for i in range(len(x)):
        print(f"P(X â‰¤ {x[i]}) = {cdf_values[i]:.4f}")

    # Step 4: Plot the CDF
    plt.figure(figsize=(10, 5))
    plt.step(x, cdf_values, where='post', label=f'Poisson CDF (Î»={lam})', color='blue')
    plt.scatter(x, cdf_values, color='red')
    plt.title('CDF of Poisson Distribution')
    plt.xlabel('k')
    plt.ylabel('P(X â‰¤ k)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
plot_poisson_cdf(lam=4, max_k=15)
```

```python
26.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np

def calculate_range_and_iqr(mean=100, std=15, size=100):
    # Step 1: Generate normal data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Step 2: Calculate Range
    data_range = np.max(data) - np.min(data)

    # Step 3: Calculate IQR
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    # Step 4: Print results
    print(f"Generated {size} samples from N({mean}, {std}Â²)")
    print(f"Range: {data_range:.2f}")
    print(f"IQR (Interquartile Range): {iqr:.2f}")
    print(f"Q1 (25th percentile): {q1:.2f}")
    print(f"Q3 (75th percentile): {q3:.2f}")

    return data

# Run the function
calculate_range_and_iqr(mean=100, std=15, size=100)
```

```python
27.Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std
    return z_scores, mean, std

def visualize_z_normalization(data):
    # Step 1: Apply Z-score normalization
    z_data, mean, std = z_score_normalization(data)

    # Step 2: Print summary
    print(f"Original Mean: {mean:.2f}, Std Dev: {std:.2f}")
    print(f"Normalized Mean: {np.mean(z_data):.2f}, Std Dev: {np.std(z_data, ddof=1):.2f}")

    # Step 3: Plot original vs normalized
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.hist(data, bins=15, color='skyblue', edgecolor='black')
    plt.title('Original Data Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.hist(z_data, bins=15, color='salmon', edgecolor='black')
    plt.title('Z-score Normalized Data Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Generate sample data (Normal distribution with mean=100, std=20)
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=20, size=100)

# Visualize the transformation
visualize_z_normalization(sample_data)
```

```python
28.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_skewness_kurtosis(mean=0, std=1, size=1000):
    # Step 1: Generate normally distributed data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Step 2: Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # By default, this is excess kurtosis (normal = 0)

    # Step 3: Print the results
    print(f"Generated {size} samples from N({mean}, {std}Â²)")
    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis (excess): {data_kurtosis:.4f}")

    # Interpretation
    print("\nInterpretation:")
    if abs(data_skewness) < 0.5:
        print("- Skewness indicates approximately symmetric data (close to 0).")
    elif data_skewness > 0:
        print("- Skewness > 0: Right (positive) skew.")
    else:
        print("- Skewness < 0: Left (negative) skew.")

    if abs(data_kurtosis) < 0.5:
        print("- Kurtosis indicates normal peakness (mesokurtic).")
    elif data_kurtosis > 0:
        print("- Kurtosis > 0: Leptokurtic (heavy tails).")
    else:
        print("- Kurtosis < 0: Platykurtic (light tails).")

# Example usage
analyze_skewness_kurtosis(mean=0, std=1, size=1000)
```
