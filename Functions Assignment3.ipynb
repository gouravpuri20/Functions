 O1. What is the difference between a function and a method in Python?
-  Here is the difference between a function and a method in Python:
Function
A function is a block of code that performs a specific task.
It is defined independently of any class.
It is called by its name.
It can have zero or more parameters (inputs).
It can return a value.
It is used for general-purpose tasks.
Method
A method is a function that is associated with an object (instance of a class).
It is defined inside a class.
It is called using the object name, followed by a dot and the method name.
It always has a first parameter, conventionally named self, which refers to the object itself.
It can access and modify the object's data (attributes).
It is used for tasks that are specific to the object.

Anonymous functions that are defined using the lambda keyword.

Functions that call themselves during their execution.

Functions defined inside other functions.

Functions that modify the behavior of other functions.

Use docstrings (triple-quoted strings) to document the purpose, parameters, and return value of a function.

6. Explain the concept of generators in Python and how they are defined.
-  In Python, generators are a type of iterator that efficiently produce a sequence of values on demand, rather than storing them all in memory at once. They are defined using a function that uses the yield keyword instead of return. This allows the function to pause and resume execution, yielding a value each time it's called, making them ideal for working with large or potentially infinite sequences.
Key Concepts:
Iterators:
Generators are a specific type of iterator, which is an object that can return the next item in a sequence one at a time.
Yield:
The yield keyword is used to produce a value from the generator. When a generator function is called, it returns a generator object that can be iterated over using next() or a loop.
Lazy Evaluation:
Generators produce values only when they are requested, making them memory-efficient, especially for large datasets or infinite sequences.
Pausing and Resuming:
Generators can be paused and resumed at the yield statement, allowing them to produce values incrementally.

7. What are the advantages of using generators over regular functions?
-  Generators offer several advantages over regular functions, primarily concerning memory efficiency and lazy evaluation. Unlike regular functions that return a complete result immediately, generators yield values one at a time, saving memory, especially when dealing with large datasets or infinite sequences. They also allow for lazy evaluation, computing values only when needed, potentially improving performance.
Here's a more detailed breakdown:
1.Memory Efficiency:
Regular functions create a complete result in memory before returning it, which can be problematic for large datasets that consume significant memory.
Generators, on the other hand, produce values one at a time on demand, minimizing memory usage.
This is particularly useful when dealing with large files or streams of data where you only need to process a portion at a time.
2.Lazy Evaluation:
Regular functions compute and return the entire result at once, even if the result isn't immediately needed.
Generators compute and yield values only when they are requested, which can be more efficient.
This is especially beneficial when dealing with operations that may be computationally expensive or where the entire result is not required.
3.Infinite Sequences and Data Pipelining:
Generators can generate an infinite sequence of values, which is not possible with regular functions.
They also enable data pipelining, where data is processed in a sequence of steps, with each generator processing a segment of data before passing it on.
This can be used to implement algorithms that require processing large datasets efficiently.
4.Improved Readability and Conciseness:
Generators can sometimes lead to more concise and readable code compared to using explicit loops or iterators.
They provide a more streamlined way to define iterative algorithms.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
-   map(), filter(), and reduce() are built-in functions in Python that facilitate functional programming paradigms. Each function operates on iterables (like lists, tuples, etc.) but serves a distinct purpose.
map()
Applies a given function to each item of an iterable.
Returns a new iterator (map object) that yi
elds the results.
Transforms each element in the iterable based on the provided function.
    numbers = [1, 2, 3, 4]
    squared = map(lambda x: x**2, numbers)
    print(list(squared))  # Output: [1, 4, 9, 16]
filter()
Filters elements from an iterable based on a given condition.
Returns a new iterator (filter object) containing only the elements for which the condition (function) is true.
Selects elements that meet a specific criterion.
    numbers = [1, 2, 3, 4, 5, 6]
    even_numbers = filter(lambda x: x % 2 == 0, numbers)
    print(list(even_numbers))  # Output: [2, 4, 6]
reduce()
Applies a function cumulatively to the items of an iterable, reducing it to a single value.
Requires importing from the functools module.
Performs a rolling computation on sequential pairs of elements.
    from functools import reduce

    numbers = [1, 2, 3, 4]
    product = reduce(lambda x, y: x * y, numbers)
    print(product)  # Output: 24
Key Differences:
map() transforms each element, filter() selects elements, and reduce() combines elements.
map() and filter() return iterators, while reduce() returns a single value.
reduce() requires importing from functools, unlike map() and filter().

```python
# Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(num):

    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def print_primes_in_range(limit):
  """
    Prints prime numbers within a specified range.

    Args:
        limit: The upper limit of the range.
  """
  for number in range(1, limit + 1):
    if is_prime(number):
      print(number)

print_primes_in_range(200)
```

```python

# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
term
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0,1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration

        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value
fib = FibonacciIterator(7)
print(list(fib) )
```

```python
#  Write a generator function in Python that yields the powers of 2 up to a given exponent
def powers_of_two(max_exponent) :
    for i in range (max_exponent + 1) :
        yield 2 ** i
print (list(powers_of_two(4)))
```

```python
# Implement a generator function that reads a file line by line and yields each line as a string
def read_file_line_by_line(file_path):
    """Reads a file line by line and yields each line as a string."""
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```
