Q2.How does broadcasting work in NumPy?

Ans Broadcasting in NumPy is a powerful feature that allows arithmetic operations between arrays of different shapes. Instead of manually replicating data to match shapes, NumPy automatically stretches the smaller array across the larger one so they are compatible for element-wise operations.

How Broadcasting Works

When operating on two arrays, NumPy compares their shapes element-wise from right to left (trailing dimensions). Two dimensions are compatible when:
1. They are equal, or

2. One of them is 1

If the dimensions donâ€™t match and none of them is 1, NumPy throws a ValueError.

Example 1: Adding a scalar to a 2D array

import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = 10

result = A + B

Broadcasting stretches B to match the shape of A:

B.shape = ()  â†’ broadcast to (2, 3)

Q6.What are the differences between NumPy arrays and Python lists?

Ans Here are the key differences between NumPy arrays and Python lists:

1. Data Type Consistency

NumPy Arrays: All elements are of the same data type (e.g., all float64 or all int32).

Python Lists: Can contain mixed data types (e.g., [1, "hello", 3.5] is valid).

2. Performance

NumPy Arrays: Much faster for numerical operations due to implementation in C and use of vectorization.

Python Lists: Slower for numeric computations; operations must loop through elements manually.
.
3. Functionality

NumPy Arrays: Support element-wise operations and broadcasting.

a = np.array([1, 2, 3])
a * 2  # [2, 4, 6]

Python Lists: Do not support element-wise operations directly.

a = [1, 2, 3]
a * 2  # [1, 2, 3, 1, 2, 3] (repeats list)

4. Memory Efficiency

NumPy Arrays: Use less memory and are more efficient.

Python Lists: Use more memory due to storing object references.

5. Advanced Features

NumPy Arrays: Support multi-dimensional arrays, slicing, reshaping, matrix operations, etc.

Python Lists: Only support basic operations; manual implementation needed for matrix or higher-dimensional data.

Q21.Why is NumPyâ€™s array slicing faster than Pythonâ€™s list slicing?

Ans NumPyâ€™s array slicing is faster than Pythonâ€™s list slicing due to several key reasons rooted in memory layout, data type uniformity, and internal optimizations:

Reasons Why NumPy Array Slicing is Faster:

1. Contiguous Memory Layout
NumPy arrays are stored in contiguous blocks of memory (like C arrays).
This enables fast access and slicing using pointer arithmetic.
In contrast, Python lists are arrays of object pointers scattered in memory.

Result: NumPy can access a slice without copying elements, just by changing the view.

2. Homogeneous Data Type
NumPy arrays contain elements of the same data type, allowing optimized, low-level operations.
Python lists can hold heterogeneous types, requiring more checks during slicing.

Result: NumPy can use vectorized operations without per-element type checking.

3. Views Instead of Copies
NumPy slicing returns a view, not a copy. This avoids unnecessary data duplication.
Python list slicing creates a new list object, copying all the sliced elements.

Result: NumPy slicing is more memory-efficient and faster.

4. Implementation in CNumPy is implemented in C, which is much faster than Pythonâ€™s native implementation.
Many array operations, including slicing, are backed by highly optimized C code.Example Comparison:
import numpy as np
import time

# NumPy array
a = np.arange(1000000)

# Python list

```python
Q11.Perform matrix multiplication using NumPy

Ans To perform matrix multiplication using NumPy

import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Method 1: Using @ operator
result1 = A @ B

# Method 2: Using np.dot()
result2 = np.dot(A, B)

# Method 3: Using np.matmul()
result3 = np.matmul(A, B)

print("Result using @ operator:\n", result1)
print("Result using np.dot():\n", result2)
print("Result using np.matmul():\n", result3)
```
