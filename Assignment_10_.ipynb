Q1.What is hypothesis testing in statistics?

Ans In hypothesis testing, one form of statistical inference, a claim about a population is evaluated using data observed from a sample of the population. The data one observes will be different depending on which individuals of the population the sample captures.

Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans Null Hypothesis (H₀)

The null hypothesis is a default or starting assumption in hypothesis testing. It usually states that there is no effect, no difference, or no relationship between variables.

It represents the status quo or what we assume to be true unless proven otherwise.

We try to disprove or reject the null hypothesis using sample data.

Example:

H₀: The average height of adult men = 170 cm

H₀: A new drug has no effect on blood pressure

Alternative Hypothesis (H₁ or Ha)

The alternative hypothesis is what you want to test or prove. It suggests that there is an effect, a difference, or a relationship.

If the evidence is strong enough to reject H₀, we accept H₁.

It is the opposite of the null hypothesis.

Example (continued):

H₁: The average height of adult men ≠ 170 cm

H₁: The new drug does affect blood pressure

Q3.What is the significance level in hypothesis testing, and why is it important?

Ans The significance level, denoted by α (alpha), is the threshold we set to decide whether to reject the null hypothesis (H₀).

It represents the maximum probability of making a Type I error, which means rejecting H₀ when it is actually true.

 Common Significance Levels:

α (Significance Level)  Meaning

0.05 (5%)   5% risk of rejecting a true H₀
0.01 (1%)   More strict → 1% risk
0.10 (10%)  More lenient → 10% risk

Why is it Important?

1. Controls the risk of false positives (Type I error)

Lower α = less chance of mistakenly rejecting H₀.

2. Sets the bar for evidence

Tells us how strong the sample evidence must be to reject H₀.

If p-value < α, the evidence is strong enough → reject H₀.

3. Impacts decision-making

In medical testing, a low α (e.g., 0.01) is often used to avoid false claims about treatment effectiveness.

In early research, a higher α (e.g., 0.10) might be acceptable for exploratory purposes.

Q4.What does a P-value represent in hypothesis testing?

Ans The P-value (short for probability value) tells you how likely it is to get a result as extreme as or more extreme than the one observed, assuming the null hypothesis (H₀) is true.

In Simple Terms:

> The P-value helps you decide whether your sample result is surprising enough to reject the null hypothesis.

Q5.How do you interpret the P-value in hypothesis testing?

Ans The P-value helps you determine whether the result of your test is statistically significant.

Basic Interpretation:

P-value ≤ α (significance level)
Reject the null hypothesis (H₀)
➤ There is strong evidence against H₀
➤ Your result is statistically significant

P-value > α
Fail to reject the null hypothesis (H₀)
➤ There is not enough evidence to support the alternative (H₁)
➤ Your result is not statistically significant.

Q6.What are Type 1 and Type 2 errors in hypothesis testing?

Ans Type I and Type II errors is crucial in hypothesis testing because it helps you evaluate the risks involved when making decisions based on sample data.

Overview

Error Type  What It Means   Real-Life Analogy

Type I Error    Rejecting a true null hypothesis    Convicting an innocent person
Type II Error   Failing to reject a false null hypothesis   Letting a guilty person go free

Type I Error (False Positive)

Definition: You reject H₀ when it is actually true.

Risk Controlled by: Significance level (α)

Commonly α = 0.05 → 5% chance of Type I error

Interpretation: You're detecting an effect or difference that doesn't actually exist.

Example:
A medical test says a person has a disease when they actually don’t.

Type II Error (False Negative)

Definition: You fail to reject H₀ when it is actually false.

Probability denoted by: β (beta)

Power of the test = 1 - β
(Higher power = lower chance of Type II error)

Interpretation: You miss a real effect or difference.

Example:
A medical test says a person does not have a disease when they actually do.

Q7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Ans Understanding the difference between one-tailed and two-tailed tests helps you choose the right hypothesis test based on what you're trying to prove.

Basic Definitions

Type of Test    What It Tests   Direction of Interest

One-tailed test Whether a value is greater than or less than a certain value    One direction only (>, <)
Two-tailed test Whether a value is different (not equal) from a certain value   Both directions (≠)

One-Tailed Test

You test for a difference in only one direction.

Use it only if you're interested in deviation in a specific direction.

Examples:

H₀: μ ≤ 100
H₁: μ > 100 (Right-tailed test)
→ You're testing if the mean is greater than 100.

H₀: μ ≥ 100
H₁: μ < 100 (Left-tailed test)
→ You're testing if the mean is less than 100.

Use when: You have a clear expectation of the direction of change.

Two-Tailed Test

You test for a difference in either direction.

This is the default when you're unsure whether the value is higher or lower.

Example:

H₀: μ = 100
H₁: μ ≠ 100
→ You're testing if the mean is different from 100 (could be higher or lower).

Use when: You’re checking for any change, not just in one direction.

Q12.What is the T-test, and how is it used in hypothesis testing?

Ans A T-test is a statistical test used to determine whether there is a significant difference between the means of two groups. It is especially useful when:
The sample size is small (typically n < 30).
The population standard deviation is unknown.
The underlying population is assumed to be normally distributed.

It uses the t-distribution, which accounts for increased variability when dealing with smaller sample sizes.

How is it used in Hypothesis Testing?

1. State the hypotheses:

Null hypothesis (H₀): There is no difference between the means.

Alternative hypothesis (H₁ or Hₐ): There is a difference.

2. Choose the significance level (α), typically 0.05.

3. Calculate the t-statistic using the appropriate formula.

4. Determine the degrees of freedom (df) and use a t-distribution table (or software) to find the critical t-value.

5. Compare the calculated t-statistic with the critical t-value, or compute the p-value:

If |t| > critical value, or p-value < α → Reject H₀.

Otherwise → Fail to reject H₀.

Q15.What is the margin of error, and how does it affect the confidence interval?

Ans The margin of error is a statistical term that quantifies the amount of random sampling error in a survey or experiment. It tells you how much you can expect your sample results to differ from the true population value.

Formula (for a sample mean when population standard deviation is known):

\text{Margin of Error (MoE)} = Z^* \times \frac{\sigma}{\sqrt{n}}

Where:

 = critical value from the Z-distribution (e.g., 1.96 for 95% confidence)

 = population standard deviation

 = sample size

If the population standard deviation is unknown and the sample size is small, you use the t-distribution instead:

\text{MoE} = t^* \times \frac{s}{\sqrt{n}}

How Does Margin of Error Affect the Confidence Interval?

A confidence interval (CI) is a range of values within which we expect the true population parameter (like the mean or proportion) to fall. The formula is:

\text{Confidence Interval} = \text{Sample Estimate} \pm \text{Margin of Error}

So, the margin of error directly determines the width of the confidence interval.

Q17.What is the Chi-square distribution, and when is it used?

Ans The Chi-square (χ²) distribution is a probability distribution that arises when you sum the squares of independent standard normal variables. In simpler terms, if you square several values that follow a standard normal distribution (mean = 0, standard deviation = 1) and then add them up, the result follows a Chi-square distribution.

It is skewed to the right, especially for small degrees of freedom.

It becomes more symmetric as the degrees of freedom increase.

It is defined only for non-negative values (χ² ≥ 0).

When Is the Chi-Square Distribution Used?

It is primarily used in statistical hypothesis testing and confidence interval estimation for categorical data. Common uses include:

1. Chi-Square Test for Independence

Purpose: To test if two categorical variables are independent.

Example: Is gender independent of voting preference?

2. Chi-Square Goodness-of-Fit Test

Purpose: To test whether a sample matches a population distribution.

Example: Does a die produce all outcomes equally?

3. Chi-Square Test for Homogeneity

Purpose: To test if different populations have the same distribution of a categorical variable.

Example: Do three different schools have the same distribution of students’ favorite subjects?

4. In Confidence Intervals for Variance

The Chi-square distribution is used when calculating confidence intervals for population variance or standard deviation, especially in normal distributions.

Q18.What is the Chi-square goodness of fit test, and how is it applied?

Ans The Chi-square goodness of fit test is a statistical hypothesis test used to determine whether a set of observed categorical data matches an expected distribution.
Here's a breakdown of how it's applied:
1. Define the Hypotheses:
Null Hypothesis (H0): The observed data follows the expected distribution.
Alternative Hypothesis (Ha): The observed data does not follow the expected distribution.
2. Determine Expected Frequencies:
For each category in your observed data, calculate the expected frequency based on the hypothesized distribution.
If no specific distribution is hypothesized, equal probabilities (e.g., a uniform distribution) are often assumed.
3. Calculate the Chi-Square Statistic:
The chi-square statistic is calculated using the following formula: χ² = Σ [(Observed - Expected)² / Expected].
Sum the results for each category.
4. Determine Degrees of Freedom:
Degrees of freedom (df) are calculated as (number of categories - 1).
5. Determine the p-value:
Using the calculated chi-square statistic and degrees of freedom, find the p-value from a chi-square distribution table or using statistical software.
The p-value represents the probability of observing a chi-square statistic as extreme as, or more extreme than, the one calculated, if the null hypothesis is true.
6. Make a Decision:
Compare the p-value to a pre-determined significance level (alpha, typically 0.05).
If the p-value is less than or equal to alpha, reject the null hypothesis and conclude that the observed distribution is significantly different from the expected distribution.
If the p-value is greater than alpha, fail to reject the null hypothesis and conclude that there is not enough evidence to say the observed distribution is significantly different from the expected distribution.
In simpler terms, the chi-square goodness-of-fit test helps you determine if the differences you see in your data are just random variation or if they point to a real difference between your sample and a specific distribution.

Q20.What is an ANOVA test, and what are its assumptions?

Ans ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more groups to determine whether at least one group mean is significantly different from the others.
ANOVA Assumptions

For ANOVA results to be valid, the following assumptions must be met:

1. Independence of observations

Data from different groups should be collected independently.

Violations can lead to misleading p-values.

2. Normality

The data in each group should be approximately normally distributed.

ANOVA is fairly robust to moderate departures from normality if sample sizes are large and equal.

3. Homogeneity of variances (homoscedasticity)

The variance among the groups should be approximately equal.

This can be tested using Levene’s test or Bartlett’s test.

Q21. What are the different types of ANOVA tests?

Ans There are several types of ANOVA tests, each suited for different experimental designs. Here's a breakdown of the main types of ANOVA and when to use each.

1. One-way ANOVA:
This is the simplest form of ANOVA and is used when you have one independent variable with three or more levels (groups) and one dependent variable.
It determines if there are any statistically significant differences between the means of the groups.
Example: Comparing the test scores of students from three different schools to see if there's a significant difference in their average performance.
2. Two-way ANOVA:
This type of ANOVA involves two or more independent variables and one dependent variable.
It can assess the main effects of each independent variable on the dependent variable, as well as any interaction effects between the independent variables.
Example: Investigating the effect of both fertilizer type (independent variable 1) and watering frequency (independent variable 2) on plant growth (dependent variable).
Other related ANOVA types and concepts:
Repeated Measures ANOVA:
Used when the same subjects are measured multiple times under different conditions.
Factorial ANOVA:
A general term for ANOVA with multiple independent variables, which includes two-way ANOVA as a specific case.
MANOVA (Multivariate ANOVA):
Used when there are multiple dependent variables.
ANCOVA (Analysis of Covariance):
Used when you want to statistically control for the effects of one or more continuous variables (covariates).
Welch's F-test ANOVA:
An alternative to the standard one-way ANOVA when the assumption of equal variances is violated.
Ranked ANOVA:
Used with ordinal data or when the assumptions of standard ANOVA are not met.
Post-hoc tests:
Used after an ANOVA to determine which specific groups have significantly different means when the ANOVA result is significant.
Understanding the different types of ANOVA tests allows researchers to select the most appropriate statistical test for their specific research question and data structure.

```python
2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Step 1: Define the range (a to b inclusive)
a = 1
b = 6
x_vals = np.arange(a, b + 1)

# Step 2: Create the discrete uniform distribution
dist = randint(a, b + 1)  # scipy.stats.randint is upper-bound exclusive

# Step 3: Calculate PMF
pmf_vals = dist.pmf(x_vals)

# Step 4: Plot the PMF
plt.figure(figsize=(8, 5))
plt.bar(x_vals, pmf_vals, color='skyblue', edgecolor='black', width=0.6)
plt.title(f'Discrete Uniform Distribution PMF (a = {a}, b = {b})')
plt.xlabel('x')
plt.ylabel('P(X = x)')
plt.xticks(x_vals)
plt.ylim(0, 1.1 * max(pmf_vals))
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculate the PDF (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome, must be 0 or 1
    p (float): The probability of success (x = 1), 0 ≤ p ≤ 1

    Returns:
    float: The probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else 1 - p

# Example usage:
print("P(X=0):", bernoulli_pdf(0, 0.3))
print("P(X=1):", bernoulli_pdf(1, 0.3))
```

```python
6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Step 1: Define range of the discrete uniform distribution
a = 1  # lower bound
b = 6  # upper bound (inclusive)
x_vals = np.arange(a - 1, b + 2)  # include some out-of-range values for visualization

# Step 2: Create discrete uniform distribution and calculate CDF
dist = randint(a, b + 1)  # scipy randint is upper-exclusive
cdf_vals = dist.cdf(x_vals)

# Step 3: Plot CDF
plt.figure(figsize=(8, 5))
plt.step(x_vals, cdf_vals, where='post', label='CDF', color='purple')
plt.scatter(x_vals, cdf_vals, color='black')

plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('x')
plt.ylabel('F(x) = P(X ≤ x)')
plt.xticks(np.arange(a - 1, b + 2))
plt.yticks(np.linspace(0, 1, b - a + 2))
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
16.Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

# Sample dataset: test scores
data = [82, 85, 88, 90, 79, 84, 83, 87, 91, 86]

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation (ddof=1)
n = len(data)

# Confidence level
confidence = 0.95
alpha = 1 - confidence

# t-critical value for two-tailed test
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Output
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Confidence Interval (95%) = ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

import numpy as np
from scipy.stats import norm

def z_test_sample_mean(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Perform a Z-test to compare a sample mean to a known population mean.

    Parameters:
        sample_data (list or array): Observed sample data
        pop_mean (float): Known population mean
        pop_std (float): Known population standard deviation
        alpha (float): Significance level (default = 0.05)

    Returns:
        dict: Test results including z-statistic, p-value, and decision
    """
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-statistic
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Interpretation
    result = "Reject H₀" if p_value < alpha else "Fail to Reject H₀"

    # Output
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z_stat:.3f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Decision at α = {alpha}: {result}")

    return {
        "sample_mean": sample_mean,
        "z_statistic": z_stat,
        "p_value": p_value,
        "decision": result
    }

# -------------------------
# Example usage:

# Sample data: weights of individuals
sample_data = [71, 74, 69, 72, 75, 73, 68, 70, 74, 72]

# Known population values
population_mean = 70
population_std = 3  # known σ

# Perform Z-test
z_test_sample_mean(sample_data, population_mean, population_std)
```

```python
2.Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.

import numpy as np
from scipy.stats import norm

# Step 1: Simulate random sample data
np.random.seed(42)
sample_data = np.random.normal(loc=77, scale=5, size=30)  # Mean=77, std=5, n=30

# Step 2: Set known population parameters
population_mean = 75
population_std = 5  # Known population standard deviation

# Step 3: Compute sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Step 4: Calculate Z-statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 5: One-tailed p-value (H1: mean > population mean)
p_value = 1 - norm.cdf(z_stat)

# Step 6: Output and interpretation
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.4f}")

# Significance level
alpha = 0.05
if p_value < alpha:
    print("✅ Reject the null hypothesis (H₀): The sample mean is significantly greater than 75.")
else:
    print("❌ Fail to reject the null hypothesis (H₀): No significant evidence the mean is greater than 75.")
```

```python
3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, pop_mean, pop_std, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test.

    Parameters:
        sample_data (list or array): The sample observations
        pop_mean (float): The known population mean (μ₀)
        pop_std (float): The known population standard deviation (σ)
        alpha (float): Significance level (default 0.05)
        alternative (str): 'two-sided', 'greater', or 'less'

    Returns:
        dict: z-statistic, p-value, and decision
    """
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    se = pop_std / np.sqrt(sample_size)

    z_stat = (sample_mean - pop_mean) / se

    # Calculate p-value based on the alternative hypothesis
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    # Make decision
    decision = "Reject H₀" if p_value < alpha else "Fail to reject H₀"

    # Print results
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z_stat:.3f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Decision at α = {alpha}: {decision}")

    return {
        'z_statistic': z_stat,
        'p_value': p_value,
        'decision': decision
    }

# -------------------------
# Example usage:

np.random.seed(42)
sample = np.random.normal(loc=102, scale=10, size=40)  # Sample data

# Known population parameters
population_mean = 100
population_std = 10

# Run the Z-test
one_sample_z_test(sample, pop_mean=population_mean, pop_std=population_std, alternative='two-sided')
```

```python
4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Sample and population parameters
np.random.seed(42)
sample = np.random.normal(loc=105, scale=10, size=50)  # Sample data

pop_mean = 100       # Population mean (μ₀)
pop_std = 10         # Population standard deviation (σ)
alpha = 0.05         # Significance level

# Step 2: Z-test calculation
sample_mean = np.mean(sample)
n = len(sample)
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed p-value

# Step 3: Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.4f}")
decision = "Reject H₀" if p_value < alpha else "Fail to reject H₀"
print(f"Decision at α = {alpha}: {decision}")

# Step 4: Visualize the Z-test and decision regions
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

z_crit = norm.ppf(1 - alpha/2)  # Critical z-value (two-tailed)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Curve', color='blue')
plt.fill_between(x, y, where=(x <= -z_crit) | (x >= z_crit), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z_stat, color='green', linestyle='--', linewidth=2, label=f'Z-statistic = {z_stat:.2f}')
plt.axvline(-z_crit, color='black', linestyle=':', label=f'Critical Z = ±{z_crit:.2f}')
plt.axvline(z_crit, color='black', linestyle=':')

plt.title('Two-Tailed Z-test Visualization')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=50, alpha=0.05):
    """
    Visualizes Type I and Type II errors for a one-tailed right-sided Z-test.

    Parameters:
        mu0 (float): Null hypothesis mean
        mu1 (float): Alternative hypothesis mean
        sigma (float): Population standard deviation
        n (int): Sample size
        alpha (float): Significance level
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value and cutoff point under H0
    z_crit = norm.ppf(1 - alpha)
    x_crit = mu0 + z_crit * se  # critical value in original units

    # Range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    h0_pdf = norm.pdf(x, mu0, se)
    h1_pdf = norm.pdf(x, mu1, se)

    # Type II error: probability of not rejecting H0 when H1 is true
    beta = norm.cdf((x_crit - mu1) / se)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, h0_pdf, label='H₀: μ = {:.1f}'.format(mu0), color='blue')
    plt.plot(x, h1_pdf, label='H₁: μ = {:.1f}'.format(mu1), color='green')

    # Shade Type I Error (α): area under H0 right of critical value
    plt.fill_between(x, h0_pdf, where=(x >= x_crit), color='red', alpha=0.3, label='Type I Error (α)')

    # Shade Type II Error (β): area under H1 left of critical value
    plt.fill_between(x, h1_pdf, where=(x <= x_crit), color='orange', alpha=0.3, label='Type II Error (β)')

    # Plot vertical line at critical value
    plt.axvline(x_crit, color='black', linestyle='--', label='Critical Value = {:.2f}'.format(x_crit))

    plt.title('Visualization of Type I and Type II Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print error probabilities
    print(f"Critical Value (sample mean cutoff): {x_crit:.2f}")
    print(f"Type I Error (α): {alpha:.4f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")
```

```python
6.Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind

# Step 1: Simulate two independent datasets
np.random.seed(42)

group1 = np.random.normal(loc=100, scale=10, size=30)  # Group 1: mean=100
group2 = np.random.normal(loc=105, scale=12, size=30)  # Group 2: mean=105

# Step 2: Perform independent T-test (Welch’s t-test: equal_var=False)
t_stat, p_value = ttest_ind(group1, group2, equal_var=False)

# Step 3: Output the results
print(f"Group 1 Mean: {np.mean(group1):.2f}")
print(f"Group 2 Mean: {np.mean(group2):.2f}")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpretation
alpha = 0.05
if p_value < alpha:
    print("✅ Reject the null hypothesis: The group means are significantly different.")
else:
    print("❌ Fail to reject the null hypothesis: No significant difference between group means.")
```

```python
7.Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Step 1: Simulate paired sample data
np.random.seed(42)
before = np.random.normal(loc=70, scale=8, size=30)
after = before + np.random.normal(loc=3, scale=5, size=30)  # assumed improvement

# Step 2: Perform paired sample T-test
t_stat, p_value = ttest_rel(before, after)

# Step 3: Print the results
print(f"Mean (Before): {np.mean(before):.2f}")
print(f"Mean (After): {np.mean(after):.2f}")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("✅ Reject H₀: There is a significant difference between before and after.")
else:
    print("❌ Fail to reject H₀: No significant difference detected.")

# Step 4: Visualize the paired data
plt.figure(figsize=(10, 5))

# Line plot for individual pairs
for i in range(len(before)):
    plt.plot([0, 1], [before[i], after[i]], color='gray', linestyle='--', marker='o')

# Mean lines
plt.plot([0]*len(before), before, 'bo', label='Before')
plt.plot([1]*len(after), after, 'go', label='After')

plt.xticks([0, 1], ['Before', 'After'])
plt.ylabel('Scores')
plt.title('Paired Sample Comparison (Before vs After)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
8.Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
from scipy.stats import norm, t

# Step 1: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=102, scale=10, size=30)  # sample from N(102, 10)
population_mean = 100       # population mean under H₀
population_std = 10         # known population standard deviation (for Z-test)

# Step 2: Sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

# Step 3: Z-test (population standard deviation known)
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
z_p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed

# Step 4: T-test (population standard deviation unknown)
t_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
t_p_value = 2 * (1 - t.cdf(abs(t_stat), df=n - 1))

# Step 5: Print results
print("===== Z-Test (σ known) =====")
print(f"Sample Mean       = {sample_mean:.2f}")
print(f"Z-statistic       = {z_stat:.3f}")
print(f"P-value (Z-test)  = {z_p_value:.4f}")

print("\n===== T-Test (σ unknown) =====")
print(f"Sample Std Dev    = {sample_std:.2f}")
print(f"T-statistic       = {t_stat:.3f}")
print(f"P-value (T-test)  = {t_p_value:.4f}")

# Step 6: Decision
alpha = 0.05
print("\n===== Conclusion (α = 0.05) =====")
print(f"Z-test: {'✅ Reject H₀' if z_p_value < alpha else '❌ Fail to reject H₀'}")
print(f"T-test: {'✅ Reject H₀' if t_p_value < alpha else '❌ Fail to reject H₀'}")
```

```python
9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import norm, t

def confidence_interval(data, confidence=0.95, pop_std=None):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default is 0.95 for 95%)
        pop_std (float): Population standard deviation (if known). If None, uses sample std and t-distribution.

    Returns:
        (lower_bound, upper_bound): Tuple with confidence interval bounds
    """
    n = len(data)
    sample_mean = np.mean(data)

    if pop_std is not None:
        # Z-distribution (population std known)
        z_score = norm.ppf(1 - (1 - confidence) / 2)
        margin_error = z_score * (pop_std / np.sqrt(n))
        dist = "Z-distribution"
    else:
        # T-distribution (population std unknown)
        sample_std = np.std(data, ddof=1)
        t_score = t.ppf(1 - (1 - confidence) / 2, df=n - 1)
        margin_error = t_score * (sample_std / np.sqrt(n))
        dist = "T-distribution"

    lower_bound = sample_mean - margin_error
    upper_bound = sample_mean + margin_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Confidence Interval ({int(confidence*100)}%) using {dist}:")
    print(f"[{lower_bound:.2f}, {upper_bound:.2f}]")

    return (lower_bound, upper_bound)
```

```python
10.Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import norm, t

def margin_of_error(data, confidence=0.95, pop_std=None):
    """
    Calculate the margin of error for a given confidence level using sample data.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default is 0.95)
        pop_std (float or None): Population standard deviation. If None, use sample std and t-distribution.

    Returns:
        float: Margin of error
    """
    n = len(data)
    mean = np.mean(data)

    if pop_std is not None:
        # Z-distribution
        se = pop_std / np.sqrt(n)
        z_score = norm.ppf(1 - (1 - confidence) / 2)
        moe = z_score * se
        dist_used = "Z-distribution (σ known)"
    else:
        # T-distribution
        sample_std = np.std(data, ddof=1)
        se = sample_std / np.sqrt(n)
        t_score = t.ppf(1 - (1 - confidence) / 2, df=n - 1)
        moe = t_score * se
        dist_used = "T-distribution (σ unknown)"

    print(f"Sample Mean: {mean:.2f}")
    print(f"Sample Size: {n}")
    print(f"Confidence Level: {int(confidence*100)}%")
    print(f"Distribution Used: {dist_used}")
    print(f"Margin of Error: ±{moe:.2f}")

    return moe
```

```python
11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior_H, likelihood_D_given_H, likelihood_D_given_not_H):
    """
    Calculates the posterior probability using Bayes' Theorem.

    Parameters:
        prior_H (float): Prior probability P(H)
        likelihood_D_given_H (float): Likelihood P(D|H)
        likelihood_D_given_not_H (float): Likelihood P(D|¬H)

    Returns:
        float: Posterior probability P(H|D)
    """
    prior_not_H = 1 - prior_H

    # Calculate total probability of evidence (P(D))
    evidence = (likelihood_D_given_H * prior_H) + (likelihood_D_given_not_H * prior_not_H)

    # Apply Bayes' Theorem
    posterior = (likelihood_D_given_H * prior_H) / evidence

    print("===== Bayesian Inference Result =====")
    print(f"P(H)        = {prior_H:.4f}  (Prior)")
    print(f"P(D|H)      = {likelihood_D_given_H:.4f}  (Likelihood)")
    print(f"P(D|¬H)     = {likelihood_D_given_not_H:.4f}")
    print(f"P(D)        = {evidence:.4f}  (Evidence)")
    print(f"P(H|D)      = {posterior:.4f}  (Posterior)")
    return posterior
```

```python
12.Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table (observed frequencies)
# Example: Gender vs. Preference
data = {
    'Like':   [30, 20],  # Male, Female
    'Dislike': [10, 40]
}
table = pd.DataFrame(data, index=['Male', 'Female'])

# Step 2: Perform Chi-Square Test for Independence
chi2, p, dof, expected = chi2_contingency(table)

# Step 3: Print the results
print("=== Chi-Square Test for Independence ===")
print("Observed Table:")
print(table)
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=table.index, columns=table.columns))

print(f"\nChi-Square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

# Step 4: Interpretation
alpha = 0.05
if p < alpha:
    print("✅ Reject H₀: There IS a significant association between the variables.")
else:
    print("❌ Fail to reject H₀: No significant association found between the variables.")
```

```python
13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

import pandas as pd
import numpy as np

def calculate_expected_frequencies(observed_table):
    """
    Calculate expected frequencies for a contingency table.

    Parameters:
        observed_table (pd.DataFrame): Observed frequencies (contingency table)

    Returns:
        pd.DataFrame: Expected frequency table
    """
    # Convert to NumPy array
    observed = observed_table.values
    row_totals = observed.sum(axis=1).reshape(-1, 1)      # reshape for matrix multiplication
    col_totals = observed.sum(axis=0).reshape(1, -1)
    grand_total = observed.sum()

    expected = (row_totals @ col_totals) / grand_total

    return pd.DataFrame(expected, index=observed_table.index, columns=observed_table.columns)

# Example usage
# Observed data: Gender vs Preference
data = {
    'Yes': [25, 15],
    'No': [10, 30]
}
observed_df = pd.DataFrame(data, index=['Male', 'Female'])

expected_df = calculate_expected_frequencies(observed_df)

print("=== Observed Frequencies ===")
print(observed_df)

print("\n=== Expected Frequencies ===")
print(expected_df.round(2))
```

```python
14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy.stats import chisquare

# Step 1: Define observed data
observed = np.array([8, 9, 12, 11, 10, 10])  # Results from die rolls

# Step 2: Define expected frequencies (uniform distribution if die is fair)
expected = np.array([10, 10, 10, 10, 10, 10])

# Step 3: Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Output results
print("=== Chi-square Goodness-of-Fit Test ===")
print(f"Observed Frequencies: {observed}")
print(f"Expected Frequencies: {expected}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Conclusion
alpha = 0.05
if p_value < alpha:
    print("✅ Reject H₀: Observed data does not fit the expected distribution.")
else:
    print("❌ Fail to reject H₀: Observed data fits the expected distribution.")
```
